```python
# Для единообразия перенес код в Ноутбук и перевел комменты на русский. Но для себя сохраню оригинальный вариант
from PIL import Image
import pytesseract
import os

DIR = 'pass_photos'                 # Путь к папке с фотографиями паспортов
LANG = 'rus'                        # Язык для распознавания текста
PSM = 6                             # Page Segmentation Mode (--psm). Больше информации в Tesseract --help
OEM = 3                             # Engine Mode (--oem). Больше информации в Tesseract --help
```

```python
# Собираем список названий фото в рабочей папке
def get_files(directory: str) -> list:
    names = []
    for filename in os.listdir(directory):
        if filename.endswith(".jpeg") or filename.endswith(".jpg") or filename.endswith(".png"):
            names.append(os.path.join(directory, filename))

    return names
```

```python
# Распознаем текст с помощью Tesseract OCR
def recognize(img_path: str, lang: str, psm: int, oem: int) -> str:

    config = f'--oem {oem} --psm {psm}'
    text_data = f"========IMAGE '{img_path}' ========\n"           # Форматируем вывод
    text_data = text_data + pytesseract.image_to_string(Image.open(img_path), lang=lang, config=config) + "\n"
    print(text_data)
    return text_data
```

```python
# Запус всего процесса
text = ""
images = get_files(directory=DIR)
for img in images:
    text = text + recognize(img_path=img, lang=LANG, psm=PSM, oem=OEM)
print(f"Count of recognized files = {len(images)}")
with open("task_1.txt", 'w') as f:
    f.write(text)
```
